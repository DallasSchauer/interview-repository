```python
## Contains Duplicate

## Given an integer array nums, return true if any value appears at least twice in the array,
## and return false if every element is distinct.

## https://leetcode.com/problems/contains-duplicate/description/ 

def containsDuplicate(nums) -> bool:
    return len(set(nums)) != len(nums)

def containsDuplicateAlternate(nums) -> bool:
    tracker = []
    
    for i in nums:
        if i not in tracker:
            tracker.append(i)
        else:
            return True
    return False
    

input1 = [1,2,3,1]
input2 = [1,2,3,4]
input3 = [1,1,1,3,3,4,3,2,4,2]

print("SOLUTION 1")
print("1 ✅") if (containsDuplicate(input1) == True) else print("1 ❌")
print("2 ✅") if (containsDuplicate(input2) == False) else print("2 ❌")
print("3 ✅") if (containsDuplicate(input3) == True) else print("3 ❌")

print("SOLUTION 2")
print("1 ✅") if (containsDuplicateAlternate(input1) == True) else print("1 ❌")
print("2 ✅") if (containsDuplicateAlternate(input2) == False) else print("2 ❌")
print("3 ✅") if (containsDuplicateAlternate(input3) == True) else print("3 ❌")
```

```python
## Two Sum

## Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.
## You may assume that each input would have exactly one solution, and you may not use the same element twice.
## You can return the answer in any order.

## https://leetcode.com/problems/two-sum/
    
def twoSum(nums, target):
        values = dict()

        for i, n in enumerate(nums):
            if (target - n) in values:
                return [values[target-n], i]
            else:
                values[n] = i

        return None
    
input1 = [2,7,11,15]
input2 = [3,2,4]
input3 = [3,3]

print("SOLUTION 1")
print("1 ✅") if (twoSum(input1, 6) == None) else print("1 ❌")
print("2 ✅") if (twoSum(input2, 6) == [1, 2]) else print("2 ❌")
print("3 ✅") if (twoSum(input3, 6) == [0, 1]) else print("3 ❌")
```

```python
## Group Anagrams

## Given an array of strings strs, group the anagrams together. You can return the answer in any order.
## An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase,
## typically using all the original letters exactly once.

## https://leetcode.com/problems/group-anagrams/

def groupAnagrams(strs):
        alphabetical = dict()
        ret = []
        retlen = 0

        for l in strs:
            sortedStr = str(sorted(l))
            if sortedStr in alphabetical:
                ret[alphabetical[sortedStr]].append(l)
            else:
                alphabetical[sortedStr] = retlen
                ret.append([])
                ret[retlen].append(l)
                retlen += 1

        return ret
    
input1 = ["eat","tea","tan","ate","nat","bat"]
input2 = [""]
input3 = ["a"]

print("SOLUTION 1")
print("1 ✅") if (groupAnagrams(input1) == [["eat","tea","ate"],["tan","nat"],["bat"]]) else print("1 ❌")
print("2 ✅") if (groupAnagrams(input2) == [[""]]) else print("2 ❌")
print("3 ✅") if (groupAnagrams(input3) == [["a"]]) else print("3 ❌")
```

```python
# Top K Frequent Elements

# Given an integer array nums and an integer k, return the k most frequent elements. You may return the answer in any order.

# https://leetcode.com/problems/top-k-frequent-elements/

def topKFrequent(nums, k):
        counts = dict()

        for n in nums:
            if n in counts:
                counts[n] += 1
            else:
                counts[n] = 1
        
        return sorted(counts, key=counts.get, reverse=True)[0:k]
    
nums1 = [1,1,1,2,2,3]
k1 = 2

nums2 = [1]
k2 = 1

print("SOLUTION 1")
print("1 ✅") if (topKFrequent(nums1, k1) == [1, 2]) else print("1 ❌")
print("2 ✅") if (topKFrequent(nums2, k2) == [1]) else print("2 ❌")
```

```python
# Product of Array Except Self

# Given an integer array nums, return an array answer such that answer[i] is equal to the product
# of all the elements of nums except nums[i]. The product of any prefix or suffix of nums is guaranteed
# to fit in a 32-bit integer. You must write an algorithm that runs in O(n) time and without using the division operation.

# https://leetcode.com/problems/product-of-array-except-self/submissions/

def productExceptSelf(nums):
        length = len(nums)
        ret = [1] * length
        pre = 1
        post = 1

        for i in range(length):
            ret[i] *= pre
            pre = pre * nums[i]
            ret[length-i-1] *= post
            post = post*nums[length-i-1]

        return ret
```
